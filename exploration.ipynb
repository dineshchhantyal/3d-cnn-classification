```python
# Quick directory exploration
def explore_data(path):
    """Quick overview of data structure"""
    p = Path(path)
    print("üìÅ Main files:")
    for item in p.iterdir():
        if item.is_file():
            print(f"   {item.name}")

    # Check subdirectories
    reg_img = p / "registered_images"
    reg_lbl = p / "registered_label_images"

    print("reg_img:", reg_img)
    print("reg_lbl:", reg_lbl)
    if reg_img.exists():
        print(f"\nüì∏ Images: {len(list(reg_img.glob('*.tif*')))} files")
    if reg_lbl.exists():
        print(f"üè∑Ô∏è  Labels: {len(list(reg_lbl.glob('*.tif*')))} files")

explore_data(data_base_path)
```

```python
def load_metadata(path):
    """Load and summarize classification data from DeathAndMitoticClass.txt"""
    p = Path(path)
    class_file = p / "DeathAndMitoticClass.txt"

    if not class_file.exists():
        print(f"‚ùå File not found: {class_file}")
        return {}

    try:
        print(f"üìÇ Loading classification data from: {class_file}")

        # Load data with pandas - handles irregular spacing
        df = pd.read_csv(
            class_file,
            sep="\s+",
            header=None,
            names=["frame", "nucleus_id", "mitotic", "death"],
        )

        print(f"‚úÖ Loaded {len(df)} classifications")
        print(f"Columns: {list(df.columns)}")

        # Show data distribution
        print(f"\nData distribution:")
        print(f"Frames: {df['frame'].min()} - {df['frame'].max()}")
        print(f"Nucleus IDs: {df['nucleus_id'].nunique()} unique")
        print(f"Mitotic events: {df['mitotic'].sum()}")
        print(f"Death events: {df['death'].sum()}")

        return {"classes": df}

    except Exception as e:
        print(f"‚ùå Error loading file: {e}")
        return {}


# Load metadata
metadata = load_metadata(data_base_path)

if "classes" in metadata:
    df = metadata["classes"]

    print("\nFirst 10 rows:")
    print(df.head(10))

    # Classification breakdown
    print(f"\nClassification breakdown:")
    print(f"Mitotic only: {len(df[(df['mitotic']==1) & (df['death']==0)])}")
    print(f"Death only: {len(df[(df['mitotic']==0) & (df['death']==1)])}")
    print(f"Both: {len(df[(df['mitotic']==1) & (df['death']==1)])}")
    print(f"Neither: {len(df[(df['mitotic']==0) & (df['death']==0)])}")
else:
    print("‚ùå Failed to load classification data")
```

```python
# Quick image visualization
def show_sample_images(path, n_samples=2):
    """Show sample images and labels"""
    p = Path(path)
    img_files = list((p / "registered_images").glob("*.tif*"))[:n_samples]
    lbl_files = list((p / "registered_label_images").glob("*.tif*"))[:n_samples]
    lbl_file_path = p / "registered_label_images"

    print(img_files[0])
    print(lbl_files[0])

    fig, axes = plt.subplots(2, n_samples, figsize=(12, 8))

    for i, img_file in enumerate(img_files):
        # Load and show image
        img = tifffile.imread(img_file)
        if len(img.shape) == 3:
            img = img[img.shape[0] // 2]  # Middle slice

        axes[0, i].imshow(img, cmap="gray")
        axes[0, i].set_title(f"Image: {img_file.stem}")
        axes[0, i].axis("off")

        # Load and show label
        lbl_file = lbl_file_path / img_file.name.replace("nuclei", "label")
        lbl = tifffile.imread(lbl_file)
        if len(lbl.shape) == 3:
            lbl = lbl[lbl.shape[0] // 2]  # Middle slice

        axes[1, i].imshow(lbl, cmap="jet")
        axes[1, i].set_title(f"Label: {lbl_file.stem}")
        axes[1, i].axis("off")

    plt.tight_layout()
    plt.show()

    return img.shape, lbl.shape


img_shape, lbl_shape = show_sample_images(data_base_path)
print(f"Image shape: {img_shape}, Label shape: {lbl_shape}")
```

```python
def show_specific_z_slice(path, frame, nucleus_id, z_slice=None):
    """Show a specific Z-slice of both raw and label images"""
    p = Path(path)
    img_dir = p / "registered_images"
    lbl_dir = p / "registered_label_images"
    
    # Find files for the specified frame
    img_files = list(img_dir.glob(f"nuclei_reg8_{frame}.tif"))
    lbl_files = list(lbl_dir.glob(f"label_reg8_{frame}.tif"))
    
    if not img_files or not lbl_files:
        print(f"‚ùå Files not found for frame {frame}")
        return
    
    # Load images
    img = tifffile.imread(img_files[0])
    lbl = tifffile.imread(lbl_files[0])
    
    print(f"üìä Image shape: {img.shape}")
    print(f"üìä Label shape: {lbl.shape}")
    print(f"üìä Available Z-slices: 0 to {img.shape[0]-1}")
    
    # If no specific slice requested, find the best slice for the nucleus
    if z_slice is None:
        print(f"üîç Finding best slice for nucleus {nucleus_id}...")
        best_slice = 0
        max_area = 0
        
        for slice_idx in range(lbl.shape[0]):
            slice_2d = lbl[slice_idx]
            nucleus_mask = slice_2d == nucleus_id
            area = np.sum(nucleus_mask)
            
            if area > max_area:
                max_area = area
                best_slice = slice_idx
        
        z_slice = best_slice
        print(f"üéØ Best slice for nucleus {nucleus_id}: {z_slice} (area: {max_area} pixels)")
    
    # Validate slice index
    if z_slice >= img.shape[0] or z_slice < 0:
        print(f"‚ùå Invalid slice {z_slice}. Must be between 0 and {img.shape[0]-1}")
        return
    
    # Extract the specific slice
    img_slice = img[z_slice]
    lbl_slice = lbl[z_slice]
    
    # Create visualization
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    
    # Raw image
    axes[0, 0].imshow(img_slice, cmap="gray")
    axes[0, 0].set_title(f"Raw Image\nFrame {frame}, Z-slice {z_slice}")
    axes[0, 0].axis("off")
    
    # All labels
    axes[0, 1].imshow(lbl_slice, cmap="tab20")
    axes[0, 1].set_title(f"All Labels\nFrame {frame}, Z-slice {z_slice}")
    axes[0, 1].axis("off")
    
    # Create color-coded labels
    colored_labels = np.zeros((*lbl_slice.shape, 3), dtype=np.uint8)
    other_nuclei_mask = (lbl_slice > 0) & (lbl_slice != nucleus_id)
    target_mask = (lbl_slice == nucleus_id)
    colored_labels[other_nuclei_mask] = [100, 150, 255]  # Light blue
    colored_labels[target_mask] = [255, 50, 50]  # Bright red
    
    axes[0, 2].imshow(colored_labels)
    axes[0, 2].set_title(f"Color-Coded Labels\nTarget {nucleus_id} (Red), Others (Blue)")
    axes[0, 2].axis("off")
    
    # Target nucleus only
    target_only = np.zeros_like(lbl_slice)
    target_only[target_mask] = nucleus_id
    axes[1, 0].imshow(target_only, cmap="Reds", vmin=0, vmax=nucleus_id+1)
    axes[1, 0].set_title(f"Target Nucleus {nucleus_id} Only\nArea: {np.sum(target_mask)} pixels")
    axes[1, 0].axis("off")
    
    # Overlay on raw image
    axes[1, 1].imshow(img_slice, cmap="gray")
    axes[1, 1].imshow(np.ma.masked_where(~target_mask, target_mask), cmap="Reds", alpha=0.7)
    axes[1, 1].set_title(f"Nucleus {nucleus_id} Overlay")
    axes[1, 1].axis("off")
    
    # Show all unique labels in this slice
    unique_labels = np.unique(lbl_slice)
    unique_labels = unique_labels[unique_labels > 0]  # Remove background
    
    # Create a text summary
    axes[1, 2].text(0.1, 0.9, f"Slice Information:", fontsize=14, fontweight='bold', transform=axes[1, 2].transAxes)
    axes[1, 2].text(0.1, 0.8, f"Frame: {frame}", fontsize=12, transform=axes[1, 2].transAxes)
    axes[1, 2].text(0.1, 0.7, f"Z-slice: {z_slice}", fontsize=12, transform=axes[1, 2].transAxes)
    axes[1, 2].text(0.1, 0.6, f"Target nucleus: {nucleus_id}", fontsize=12, transform=axes[1, 2].transAxes)
    axes[1, 2].text(0.1, 0.5, f"Target area: {np.sum(target_mask)} pixels", fontsize=12, transform=axes[1, 2].transAxes)
    axes[1, 2].text(0.1, 0.4, f"Total nuclei: {len(unique_labels)}", fontsize=12, transform=axes[1, 2].transAxes)
    
    # Show first few nucleus IDs
    nucleus_list = list(unique_labels[:10])
    axes[1, 2].text(0.1, 0.3, f"Nucleus IDs: {nucleus_list}{'...' if len(unique_labels) > 10 else ''}", 
                    fontsize=10, transform=axes[1, 2].transAxes)
    
    axes[1, 2].text(0.1, 0.2, f"Image intensity: {img_slice.min():.1f} - {img_slice.max():.1f}", 
                    fontsize=10, transform=axes[1, 2].transAxes)
    
    target_present = nucleus_id in unique_labels
    status = "‚úÖ PRESENT" if target_present else "‚ùå ABSENT"
    axes[1, 2].text(0.1, 0.1, f"Target status: {status}", 
                    fontsize=12, fontweight='bold', 
                    color='green' if target_present else 'red',
                    transform=axes[1, 2].transAxes)
    
    axes[1, 2].set_xlim(0, 1)
    axes[1, 2].set_ylim(0, 1)
    axes[1, 2].axis("off")
    
    plt.suptitle(f"Z-Slice Analysis - Frame {frame}, Slice {z_slice}\n"
                 f"Files: nuclei_reg8_{frame}.tif & label_reg8_{frame}.tif", 
                 fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Print detailed statistics
    print(f"\nüìä Detailed Z-Slice Analysis:")
    print(f"   üóÇÔ∏è  Files: nuclei_reg8_{frame}.tif & label_reg8_{frame}.tif")
    print(f"   üìê Z-slice: {z_slice} of {img.shape[0]-1} (0-indexed)")
    print(f"   üéØ Target nucleus {nucleus_id}: {'FOUND' if target_present else 'NOT FOUND'}")
    if target_present:
        print(f"   üìè Target area: {np.sum(target_mask)} pixels")
        if np.sum(target_mask) > 0:
            y_coords, x_coords = np.where(target_mask)
            centroid = (int(np.mean(x_coords)), int(np.mean(y_coords)))
            print(f"   üìç Target centroid: {centroid}")
    print(f"   üî¢ Total nuclei in slice: {len(unique_labels)}")
    print(f"   üè∑Ô∏è  All nucleus IDs: {list(unique_labels)}")
    print(f"   üñºÔ∏è  Image dimensions: {img_slice.shape}")
    print(f"   üåà Intensity range: {img_slice.min():.1f} - {img_slice.max():.1f}")

# Show specific slice (e.g., slice 5) for nucleus 1 in frame 25
# show_specific_z_slice(data_base_path, frame=25, nucleus_id=1, z_slice=5)
```

```python
def extract_nucleus_crops(path, metadata, crop_size=64):
    """Extract crops around nuclei using label images to find positions"""
    if "classes" not in metadata:
        print("No classification data found")
        return []

    df = metadata["classes"]
    p = Path(path)
    img_dir = p / "registered_images"
    lbl_dir = p / "registered_label_images"
    crops = []

    # Process first 5 events
    for idx, row in df.head(5).iterrows():
        try:
            frame = int(row["frame"])
            nucleus_id = int(row["nucleus_id"])
            is_mitotic = int(row["mitotic"])
            is_death = int(row["death"])

            print(f"Processing nucleus {nucleus_id} at frame {frame}")

            # Find label file for this frame - use exact naming pattern
            lbl_files = list(lbl_dir.glob(f"label_reg8_{frame}.tif"))
            if not lbl_files:
                print(f"  ‚ùå No label file for frame {frame}")
                continue

            # Load label image to find nucleus position
            lbl_img = tifffile.imread(lbl_files[0])
            print(f"  üìä Label shape: {lbl_img.shape}")

            # Search through all slices to find the nucleus
            nucleus_found = False
            best_slice = 0
            max_area = 0

            for slice_idx in range(lbl_img.shape[0]):
                slice_2d = lbl_img[slice_idx]
                nucleus_mask = slice_2d == nucleus_id
                area = np.sum(nucleus_mask)

                if area > max_area:
                    max_area = area
                    best_slice = slice_idx
                    nucleus_found = True

            if not nucleus_found:
                unique_ids = np.unique(lbl_img)
                print(
                    f"  ‚ùå Nucleus {nucleus_id} not found. Available IDs: {unique_ids[:20]}..."
                )
                continue

            print(
                f"  üîç Found nucleus {nucleus_id} in slice {best_slice} with area {max_area}"
            )

            # Use the best slice
            lbl_2d = lbl_img[best_slice]
            nucleus_mask = lbl_2d == nucleus_id

            # Get centroid
            y_coords, x_coords = np.where(nucleus_mask)
            y_center = int(np.mean(y_coords))
            x_center = int(np.mean(x_coords))

            print(f"  üìç Position: ({x_center}, {y_center})")

            # Extract time series around event
            img_time_series = []
            lbl_time_series = []

            for t in range(max(1, frame - 2), frame + 3):  # 5 frames
                print(f"    Looking for frame {t}...")

                # Find image files - use exact naming pattern
                img_files = list(img_dir.glob(f"nuclei_reg8_{t}.tif"))
                if not img_files:
                    print(f"    ‚ùå No image file for frame {t}")
                    continue

                # Find label files - use exact naming pattern
                lbl_files_t = list(lbl_dir.glob(f"label_reg8_{t}.tif"))
                if not lbl_files_t:
                    print(f"    ‚ùå No label file for frame {t}")
                    continue

                print(f"    ‚úÖ Found: {img_files[0].name} <-> {lbl_files_t[0].name}")

                # Load and crop image
                img = tifffile.imread(img_files[0])
                if len(img.shape) == 3:
                    img = img[best_slice]

                # Load and crop label
                lbl = tifffile.imread(lbl_files_t[0])
                if len(lbl.shape) == 3:
                    lbl = lbl[best_slice]

                # Extract crops using same coordinates
                x1 = max(0, x_center - crop_size // 2)
                x2 = min(img.shape[1], x_center + crop_size // 2)
                y1 = max(0, y_center - crop_size // 2)
                y2 = min(img.shape[0], y_center + crop_size // 2)

                img_crop = img[y1:y2, x1:x2]
                lbl_crop = lbl[y1:y2, x1:x2]

                # Pad to consistent size
                padded_img_crop = np.zeros((crop_size, crop_size), dtype=img.dtype)
                padded_lbl_crop = np.zeros((crop_size, crop_size), dtype=lbl.dtype)

                h, w = img_crop.shape
                padded_img_crop[:h, :w] = img_crop
                padded_lbl_crop[:h, :w] = lbl_crop

                img_time_series.append(padded_img_crop)
                lbl_time_series.append(padded_lbl_crop)

            if len(img_time_series) >= 3:
                # Create proper labels
                if is_mitotic and is_death:
                    label, label_code = "mitotic_death", 2
                elif is_mitotic:
                    label, label_code = "mitotic", 1
                elif is_death:
                    label, label_code = "death", 3
                else:
                    label, label_code = "normal", 0

                crops.append(
                    {
                        "nucleus_id": nucleus_id,
                        "frame": frame,
                        "slice": best_slice,
                        "position": (x_center, y_center),
                        "img_time_series": np.array(img_time_series),
                        "lbl_time_series": np.array(lbl_time_series),
                        "label": label,
                        "label_code": label_code,
                    }
                )

                print(f"  ‚úÖ Extracted {len(img_time_series)} frames for {label}")
            else:
                print(f"  ‚ùå Not enough frames found ({len(img_time_series)})")

        except Exception as e:
            print(f"  ‚ùå Error: {e}")
            import traceback

            traceback.print_exc()

    return crops


# Extract crops
crops = extract_nucleus_crops(data_base_path, metadata)
print(f"\nüéØ Extracted {len(crops)} nucleus time series")
```

```python
def show_crops_with_labels(crops, n_show=2):
    """Show time series with both image and label crops with detailed metadata"""
    for i, crop in enumerate(crops[:n_show]):
        n_frames = len(crop["img_time_series"])

        # Print detailed metadata for this crop
        print(f"\n{'='*60}")
        print(f"üî¨ CROP {i+1}: Nucleus {crop['nucleus_id']} - {crop['label'].upper()}")
        print(f"{'='*60}")
        print(f"üìç Event Frame: {crop['frame']}")
        print(f"üß¨ Best Slice: {crop['slice']} (Z-depth in 3D stack)")
        print(f"üìê Position: ({crop['position'][0]}, {crop['position'][1]}) in slice")
        print(
            f"‚è±Ô∏è  Time Series: {n_frames} frames (T={-(n_frames//2)} to T={n_frames//2})"
        )
        print(f"üè∑Ô∏è  Classification: {crop['label']} (code: {crop['label_code']})")

        # Show time series
        fig, axes = plt.subplots(2, n_frames, figsize=(4 * n_frames, 8))
        if n_frames == 1:
            axes = axes.reshape(-1, 1)

        for t in range(n_frames):
            time_offset = t - (n_frames // 2)
            actual_frame = crop["frame"] + time_offset

            # Show image crop
            axes[0, t].imshow(crop["img_time_series"][t], cmap="gray")
            axes[0, t].set_title(f"Image T={time_offset}\nFrame {actual_frame}")
            axes[0, t].axis("off")

            # Add frame border for event frame
            if t == n_frames // 2:
                for spine in axes[0, t].spines.values():
                    spine.set_edgecolor("red")
                    spine.set_linewidth(3)

            # Show label crop
            axes[1, t].imshow(crop["lbl_time_series"][t], cmap="jet")
            axes[1, t].set_title(f"Label T={time_offset}\nFrame {actual_frame}")
            axes[1, t].axis("off")

            # Add frame border for event frame
            if t == n_frames // 2:
                for spine in axes[1, t].spines.values():
                    spine.set_edgecolor("red")
                    spine.set_linewidth(3)

            # Show unique label values in this crop
            unique_labels = np.unique(crop["lbl_time_series"][t])
            unique_labels = unique_labels[unique_labels > 0]  # Remove background
            print(
                f"    T={time_offset}: Found nuclei IDs {list(unique_labels[:5])}{'...' if len(unique_labels) > 5 else ''}"
            )

        # Highlight event frame
        if n_frames >= 3:
            event_idx = n_frames // 2
            axes[0, event_idx].set_title(
                f"IMAGE T=0 (EVENT)\nFrame {crop['frame']}",
                color="red",
                fontweight="bold",
                fontsize=12,
            )
            axes[1, event_idx].set_title(
                f"LABEL T=0 (EVENT)\nFrame {crop['frame']}",
                color="red",
                fontweight="bold",
                fontsize=12,
            )

        plt.suptitle(
            f"Nucleus {crop['nucleus_id']} Time Series - {crop['label'].upper()}\n"
            f"Slice {crop['slice']}, Position ({crop['position'][0]}, {crop['position'][1]})",
            fontsize=14,
            fontweight="bold",
        )
        plt.tight_layout()
        plt.show()

        # Show detailed analysis of event frame
        if n_frames >= 3:
            event_idx = n_frames // 2
            fig, axes = plt.subplots(1, 4, figsize=(20, 5))

            # Original image
            axes[0].imshow(crop["img_time_series"][event_idx], cmap="gray")
            axes[0].set_title(
                f"Raw Image\nFrame {crop['frame']}, Slice {crop['slice']}"
            )
            axes[0].axis("off")

            # Label overlay
            axes[1].imshow(crop["lbl_time_series"][event_idx], cmap="jet")
            axes[1].set_title(
                f"All Labels\nFrame {crop['frame']}, Slice {crop['slice']}"
            )
            axes[1].axis("off")

            # Highlight target nucleus only
            nucleus_id = crop["nucleus_id"]
            highlight = crop["lbl_time_series"][event_idx] == nucleus_id
            target_mask = np.zeros_like(crop["lbl_time_series"][event_idx])
            target_mask[highlight] = nucleus_id

            axes[2].imshow(target_mask, cmap="Reds", vmin=0, vmax=nucleus_id + 1)
            axes[2].set_title(
                f"Target Nucleus {nucleus_id}\nArea: {np.sum(highlight)} pixels"
            )
            axes[2].axis("off")

            # Overlay on original image
            axes[3].imshow(crop["img_time_series"][event_idx], cmap="gray")
            axes[3].imshow(
                np.ma.masked_where(~highlight, highlight), cmap="Reds", alpha=0.6
            )
            axes[3].set_title(
                f"Nucleus {nucleus_id} Overlay\n{crop['label'].upper()} Event"
            )
            axes[3].axis("off")

            plt.suptitle(
                f"Event Frame Analysis - Nucleus {nucleus_id}\n"
                f"File: nuclei_reg8_{crop['frame']}.tif & label_reg8_{crop['frame']}.tif",
                fontsize=14,
                fontweight="bold",
            )
            plt.tight_layout()
            plt.show()

            # Print nucleus statistics
            print(f"üìä Nucleus {nucleus_id} Statistics:")
            print(f"   ‚Ä¢ Area: {np.sum(highlight)} pixels")
            print(f"   ‚Ä¢ Centroid: ({crop['position'][0]}, {crop['position'][1]})")
            print(
                f"   ‚Ä¢ Slice: {crop['slice']}/{crop['img_time_series'].shape[0] if len(crop['img_time_series'].shape) > 2 else 'N/A'}"
            )
            print(
                f"   ‚Ä¢ Image intensity range: {crop['img_time_series'][event_idx].min():.1f} - {crop['img_time_series'][event_idx].max():.1f}"
            )

            # Check if nucleus is present in neighboring frames
            nucleus_presence = []
            for t in range(n_frames):
                has_nucleus = np.any(crop["lbl_time_series"][t] == nucleus_id)
                nucleus_presence.append(has_nucleus)
                time_offset = t - (n_frames // 2)
                status = "‚úÖ" if has_nucleus else "‚ùå"
                print(
                    f"   ‚Ä¢ T={time_offset}: {status} Nucleus {nucleus_id} {'present' if has_nucleus else 'absent'}"
                )


if crops:
    show_crops_with_labels(crops)
```

```python
def extract_nucleus_crops(path, metadata, crop_size=64):
    """Extract crops around nuclei using label images to find positions"""
    if "classes" not in metadata:
        print("No classification data found")
        return []

    df = metadata["classes"]
    p = Path(path)
    img_dir = p / "registered_images"
    lbl_dir = p / "registered_label_images"
    crops = []

    # Process first 5 events
    for idx, row in df.head(5).iterrows():
        try:
            frame = int(row["frame"])
            nucleus_id = int(row["nucleus_id"])
            is_mitotic = int(row["mitotic"])
            is_death = int(row["death"])

            print(f"Processing nucleus {nucleus_id} at frame {frame}")

            # Find label file for this frame - use exact naming pattern
            lbl_files = list(lbl_dir.glob(f"label_reg8_{frame}.tif"))
            if not lbl_files:
                print(f"  ‚ùå No label file for frame {frame}")
                continue

            # Load label image to find nucleus position
            lbl_img = tifffile.imread(lbl_files[0])
            print(f"  üìä Label shape: {lbl_img.shape}")

            # Search through all slices to find the nucleus
            nucleus_found = False
            best_slice = 0
            max_area = 0
            reference_position = None

            for slice_idx in range(lbl_img.shape[0]):
                slice_2d = lbl_img[slice_idx]
                nucleus_mask = slice_2d == nucleus_id
                area = np.sum(nucleus_mask)

                if area > max_area:
                    max_area = area
                    best_slice = slice_idx
                    nucleus_found = True
                    # Get reference centroid for tracking
                    y_coords, x_coords = np.where(nucleus_mask)
                    reference_position = (int(np.mean(x_coords)), int(np.mean(y_coords)))

            if not nucleus_found:
                unique_ids = np.unique(lbl_img)
                print(f"  ‚ùå Nucleus {nucleus_id} not found. Available IDs: {unique_ids[:20]}...")
                continue

            print(f"  üîç Found nucleus {nucleus_id} in slice {best_slice} with area {max_area}")
            print(f"  üìç Reference position: {reference_position}")

            # Extract time series around event
            img_time_series = []
            lbl_time_series = []
            slice_info = []  # Track which slice is used for each frame

            for t in range(max(1, frame - 2), frame + 3):  # 5 frames
                print(f"    Looking for frame {t}...")

                # Find image files - use exact naming pattern
                img_files = list(img_dir.glob(f"nuclei_reg8_{t}.tif"))
                if not img_files:
                    print(f"    ‚ùå No image file for frame {t}")
                    continue

                # Find label files - use exact naming pattern
                lbl_files_t = list(lbl_dir.glob(f"label_reg8_{t}.tif"))
                if not lbl_files_t:
                    print(f"    ‚ùå No label file for frame {t}")
                    continue

                print(f"    ‚úÖ Found: {img_files[0].name} <-> {lbl_files_t[0].name}")

                # Load images
                img = tifffile.imread(img_files[0])
                lbl = tifffile.imread(lbl_files_t[0])

                # Find nucleus in this frame across all slices
                best_slice_t = best_slice  # Start with reference slice
                best_position_t = reference_position
                nucleus_found_t = False
                min_distance = float('inf')
                
                # Search in all Z-slices for this frame
                for slice_idx in range(lbl.shape[0]):
                    slice_2d = lbl[slice_idx]
                    nucleus_mask = slice_2d == nucleus_id
                    area = np.sum(nucleus_mask)
                    
                    if area > 0:  # Nucleus found in this slice
                        # Calculate centroid
                        y_coords, x_coords = np.where(nucleus_mask)
                        current_position = (int(np.mean(x_coords)), int(np.mean(y_coords)))
                        
                        # Calculate distance from reference position
                        distance = np.sqrt((current_position[0] - reference_position[0])**2 + 
                                         (current_position[1] - reference_position[1])**2)
                        
                        if distance < min_distance:
                            min_distance = distance
                            best_slice_t = slice_idx
                            best_position_t = current_position
                            nucleus_found_t = True

                if nucleus_found_t:
                    print(f"      üéØ Found nucleus {nucleus_id} in slice {best_slice_t} at {best_position_t} (distance: {min_distance:.1f})")
                    
                    # Use the found slice and position
                    img_slice = img[best_slice_t] if len(img.shape) == 3 else img
                    lbl_slice = lbl[best_slice_t] if len(lbl.shape) == 3 else lbl
                    
                    x_center, y_center = best_position_t
                    
                else:
                    print(f"      ‚ùå Nucleus {nucleus_id} not found in any slice, using reference slice {best_slice}")
                    # Fallback to reference slice and position
                    img_slice = img[best_slice] if len(img.shape) == 3 else img
                    lbl_slice = lbl[best_slice] if len(lbl.shape) == 3 else lbl
                    x_center, y_center = reference_position

                # Extract crops using current position
                x1 = max(0, x_center - crop_size // 2)
                x2 = min(img_slice.shape[1], x_center + crop_size // 2)
                y1 = max(0, y_center - crop_size // 2)
                y2 = min(img_slice.shape[0], y_center + crop_size // 2)

                img_crop = img_slice[y1:y2, x1:x2]
                lbl_crop = lbl_slice[y1:y2, x1:x2]

                # Pad to consistent size
                padded_img_crop = np.zeros((crop_size, crop_size), dtype=img.dtype)
                padded_lbl_crop = np.zeros((crop_size, crop_size), dtype=lbl.dtype)

                h, w = img_crop.shape
                padded_img_crop[:h, :w] = img_crop
                padded_lbl_crop[:h, :w] = lbl_crop

                img_time_series.append(padded_img_crop)
                lbl_time_series.append(padded_lbl_crop)
                slice_info.append({
                    'frame': t,
                    'slice': best_slice_t if nucleus_found_t else best_slice,
                    'position': (x_center, y_center),
                    'found': nucleus_found_t,
                    'distance': min_distance if nucleus_found_t else None
                })

            if len(img_time_series) >= 3:
                # Create proper labels
                if is_mitotic and is_death:
                    label, label_code = "mitotic_death", 2
                elif is_mitotic:
                    label, label_code = "mitotic", 1
                elif is_death:
                    label, label_code = "death", 3
                else:
                    label, label_code = "normal", 0

                crops.append({
                    "nucleus_id": nucleus_id,
                    "frame": frame,
                    "slice": best_slice,  # Reference slice
                    "position": reference_position,  # Reference position
                    "img_time_series": np.array(img_time_series),
                    "lbl_time_series": np.array(lbl_time_series),
                    "label": label,
                    "label_code": label_code,
                    "slice_info": slice_info,  # Detailed tracking info
                })

                print(f"  ‚úÖ Extracted {len(img_time_series)} frames for {label}")
                
                # Print tracking summary
                print(f"  üìä Tracking summary:")
                for info in slice_info:
                    status = "üéØ FOUND" if info['found'] else "‚ùå MISSING"
                    dist_str = f"(dist: {info['distance']:.1f})" if info['distance'] is not None else ""
                    print(f"    Frame {info['frame']}: {status} in slice {info['slice']} at {info['position']} {dist_str}")
                    
            else:
                print(f"  ‚ùå Not enough frames found ({len(img_time_series)})")

        except Exception as e:
            print(f"  ‚ùå Error: {e}")
            import traceback
            traceback.print_exc()

    return crops

# Extract crops with improved tracking
crops = extract_nucleus_crops(data_base_path, metadata)
print(f"\nüéØ Extracted {len(crops)} nucleus time series")
```

```python
def combine_data(files, file_name):
    """Combine data from multiple directories into a single DataFrame."""
    combined_df = pd.DataFrame()

    for file in files:
        file_path = Path(file) / file_name
        if file_path.exists():
            try:
                print(f"üìÇ Loading data from: {file_path}")
                # Load data with pandas
                df = pd.read_csv(
                    file_path,
                    sep="\s+",
                    header=None,
                    names=["frame", "nucleus_id", "mitotic", "death"],
                )
                df["source"] = file  # Add source column to track origin
                combined_df = pd.concat([combined_df, df], ignore_index=True)
                print(f"‚úÖ Loaded {len(df)} rows from {file_path}")
            except Exception as e:
                print(f"‚ùå Error loading file {file_path}: {e}")
        else:
            print(f"‚ùå File not found: {file_path}")

    print(f"\nüéØ Combined DataFrame contains {len(combined_df)} rows")
    return combined_df


# Combine data from all files
files = [file1, file2, file3]
combined_df = combine_data(files, file_name)

# Display the first few rows of the combined DataFrame
print(combined_df.head())
```
